```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
import pandas as pd
bangalore_df = pd.read_csv("bangalore_df.csv")
```

```python
bangalore_df = bangalore_df[bangalore_df['Price (in lakhs)'] < 200]
```

```python
plt.figure(figsize=(10, 6))
plt.hexbin(bangalore_df['Floor (Ratio)'], bangalore_df['Price (in lakhs)'], gridsize=40, cmap='Reds', bins='log')
plt.colorbar(label='log10(N)')
plt.xlabel('Floor (Ratio)')
plt.ylabel('Price')
plt.title('Hexbin Density Plot: Floor (Ratio) vs Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=bangalore_df, x='Floor (Ratio)', bins=20, color='teal', edgecolor='black')
plt.title('Distribution of Floor (Ratio) in Bangalore')
plt.xlabel('Floor (Ratio)')
plt.ylabel('Number of Listings')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hexbin(bangalore_df['CurrentFloor'], bangalore_df['Price (in lakhs)'], gridsize=40, cmap='Reds', bins='log')
plt.colorbar(label='log10(N)')
plt.xlabel('CurrentFloor')
plt.ylabel('Price')
plt.title('Hexbin Density Plot: Floor (Ratio) vs Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=bangalore_df, x='CurrentFloor', bins=range(0, int(bangalore_df['CurrentFloor'].max()) + 2), color='mediumseagreen', edgecolor='black')
plt.title('Distribution of Current Floor Numbers in Bangalore')
plt.xlabel('Current Floor')
plt.ylabel('Number of Listings')
plt.grid(True)
plt.show()
```

```python
bangalore_df_furnishing = bangalore_df[bangalore_df['Furnishing'] != -1]
```

```python
bangalore_df_furnishing_copy = bangalore_df_furnishing.copy()

# Map the Furnishing codes to labels
furnishing_map = {
    0: 'Unfurnished',
    1: 'Semi-Furnished',
    2: 'Furnished'
}
bangalore_df_furnishing_copy['Furnishing'] = bangalore_df_furnishing_copy['Furnishing'].map(furnishing_map)
```

```python
plt.figure(figsize=(10, 6))
sns.boxenplot(data=bangalore_df_furnishing_copy, x='Furnishing', y='Price (in lakhs)')
plt.title('Property Price Distribution by Furnishing Type (Boxen Plot)')
plt.xlabel('Furnishing Type')
plt.ylabel('Price (in lakhs)')

plt.show()
```

```python
# Plot using FacetGrid
plt.figure(figsize=(10, 6))
g = sns.FacetGrid(bangalore_df_furnishing_copy, col="Furnishing", col_wrap=3, height=4)
g.map(sns.histplot, "Price (in lakhs)", bins=30, kde=True)
g.set_titles("{col_name}")
g.fig.subplots_adjust(top=0.9)
plt.xlim(0,150)
g.fig.suptitle("Price Distribution per Furnishing Type (Price â‰¤ 200 lakhs)", fontsize=9)
plt.show()
```

```python
facing_map = {
    0: "N", 45: "NE", 90: "E", 135: "SE",
    180: "S", 225: "SW", 270: "W", 315: "NW"
}
bangalore_df_facing = bangalore_df.copy()
bangalore_df_facing['facing label'] = bangalore_df['facing'].map(facing_map)
```

```python
plt.figure(figsize=(14, 6))
sns.violinplot(
    data=bangalore_df_facing,
    x='facing label',
    y='Price (in lakhs)',
    hue='facing',  # assign x variable to hue
    legend=False,
    palette='pastel',
    density_norm='width',  # replaces 'scale'
    inner='quartile'
)
plt.grid(True)
plt.title('Price Distribution by Facing Direction')
plt.xlabel('Facing Direction')
plt.ylabel('Price (in lakhs)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mapping 0/1 to meaningful labels
feature_map = {0: 'No', 1: 'Yes'}

# Create a copy with labels
bangalore_df_plot = bangalore_df.copy()
bangalore_df_plot['Garden'] = bangalore_df_plot['Garden/Park'].map(feature_map)
bangalore_df_plot['Main Road'] = bangalore_df_plot['Main Road'].map(feature_map)
bangalore_df_plot['Pool'] = bangalore_df_plot['Pool'].map(feature_map)

# Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Garden
sns.boxplot(data=bangalore_df_plot, x='Garden', y='Price (in lakhs)', palette='pastel', legend = False ,ax=axes[0])
axes[0].set_title('Price vs Garden')

# Main Road
sns.boxplot(data=bangalore_df_plot, x='Main Road', y='Price (in lakhs)', palette='Set2', legend = False, ax=axes[1])
axes[1].set_title('Price vs Main Road')

# Pool
sns.boxplot(data=bangalore_df_plot, x='Pool', y='Price (in lakhs)', palette='coolwarm', legend = False ,  ax=axes[2])
axes[2].set_title('Price vs Pool')

# General formatting
for ax in axes:
    ax.set_xlabel('')
    ax.set_ylabel('Price (in lakhs)')

plt.suptitle('Property Price Distribution Based on Amenities', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(data=bangalore_df_plot, x='Price (in lakhs)', hue='Garden', element='step', common_norm=False, ax=axes[0])
axes[0].set_title("Price Distribution by Garden")
axes[0].set_xlim(0, 140)
sns.histplot(data=bangalore_df_plot, x='Price (in lakhs)', hue='Main Road', element='step', common_norm=False, ax=axes[1])
axes[1].set_title("Price Distribution by Main Road")
axes[1].set_xlim(0, 140)
sns.histplot(data=bangalore_df_plot, x='Price (in lakhs)', hue='Pool', element='step', common_norm=False, ax=axes[2])
axes[2].set_title("Price Distribution by Pool")
axes[2].set_xlim(0, 140)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.barplot(data=bangalore_df_plot, x='Garden', y='Price (in lakhs)', estimator='mean', ax=axes[0])
axes[0].set_title("Average Price by Garden")

sns.barplot(data=bangalore_df_plot, x='Main Road', y='Price (in lakhs)', estimator='mean', ax=axes[1])
axes[1].set_title("Average Price by Main Road")

sns.barplot(data=bangalore_df_plot, x='Pool', y='Price (in lakhs)', estimator='mean', ax=axes[2])
axes[2].set_title("Average Price by Pool")

plt.tight_layout()
plt.show()
```
